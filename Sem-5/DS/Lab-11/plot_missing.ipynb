```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_name = 'landslide_data3_miss.csv'
df = pd.read_csv(file_name)
```

```python
column_names = pd.Series(df.columns)
missing_count = pd.Series([df[[column]].isna().sum()[0] for column in column_names])
```

```python
plt.figure(figsize=(10, 5))
plt.plot(column_names, missing_count)
plt.xlabel("Attributes")
plt.ylabel("Missing count")
plt.title("Linear plot showing attributes and their missing counts")
plt.show()
```
